```python
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
review = "The storyline was amazing, but the ending was disappointing."
result = sentiment_pipeline(review)
print(result)
```

```python
sentiment_pipeline = pipeline("sentiment-analysis", top_k=None)
result = sentiment_pipeline(review)
print(result)
```
