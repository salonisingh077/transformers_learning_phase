```python
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
review = "The storyline was amazing, but the ending was disappointing."
result = sentiment_pipeline(review)
print(result)
```

```python
sentiment_pipeline = pipeline("sentiment-analysis", top_k=None)
result = sentiment_pipeline(review)
print(result)
```

```python
from transformers import pipeline #now continuing with example 2 to fulfill requirement of multilanguage sentiment analysis
```

```python
multilingual_pipeline = pipeline( "sentiment-analysis", model="tabularisai/multilingual-sentiment-analysis")
```

```python
english_review = "The product quality is excellent!"
hindi_review = "यह उत्पाद बहुत खराब है।"
spanish_review = "El servicio fue increíble."
print(multilingual_pipeline(english_review))
print(multilingual_pipeline(hindi_review))
print(multilingual_pipeline(spanish_review))
```
